```
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
import pandas_datareader.data as web

style.use('ggplot')
```

```
df = web.DataReader('GC=F', data_source='yahoo',start='10-10-2019', end='12-06-2020')
#Creating 100 days' moving average column
df['100ma'] = df['Adj Close'].rolling(window=100).mean()
df.dropna(inplace=True)#remove the NAAs
df['100ma'].tail()
```

```
ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)

ax1.plot(df.index, df['Close'])
ax1.plot(df.index, df['100ma'])
ax2.bar(df.index, df['Volume'])

plt.show()
```

```
dfr = web.DataReader('GC=F', data_source='yahoo',start='10-10-2019', end='12-06-2020')
#Creating 50 days' moving average column
dfr['50ma'] = dfr['Adj Close'].rolling(window=50).mean()
dfr.dropna(inplace=True)#remove the NAAs
dfr['50ma'].tail()

ax1 = plt.subplot2grid((6,1), (0,0), rowspan=5, colspan=1)
ax2 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)

ax1.plot(dfr.index, dfr['Close'])
ax1.plot(dfr.index, dfr['50ma'])
ax2.bar(dfr.index, dfr['Volume'])

ax1.set(ylim=(1700, 1700))
plt.show()
```
